```python
def is_anagram(string_a:str, string_b:str) -> bool:
    """Checks if point is in circle or not.

    Arguments:
        string_a (str): First string to check
        string_b (str): Second string to check

    Returns:
        Boolean term (bool)

    """
    string_a = string_a.replace(" ", "")
    string_a = string_a.lower()
    string_a = sorted(string_a)
    string_a = "".join(string_a)

    string_b = string_b.replace(" ", "")
    string_b = string_b.lower()
    string_b = sorted(string_b)
    string_b = "".join(string_b)

    return string_a == string_b
```

```python
is_anagram("Listen", "Silent")
```
