```python
def point_hits_circle(x0:int=0, y0:int=0, r:int=1, x:int=1, y:int=1) -> bool:
    """Checks if point is in circle or not.

    Arguments:
        x0 (int): X-Coordinate of circle center
        y0 (int): Y-Coordinate of circle center
        r (int) Circle radius
        x (int): X-Coordinate of point
        y (int): Y-Coordinate of point

    Returns:
        Boolean Expressen (bool)
    """
    rp = ((x - x0) ** 2 + (y - y0) ** 2) ** 0.5
    return rp <= r
```

```python
point_hits_circle(x=0.5, y=0.5)
```
