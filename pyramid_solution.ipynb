```python
def pyramid(char:str) -> None:
    """Creates a character pyramid.

    Arguments:
        char (str): Character that is used for construction

    Returns:
        None
    """

    alphabet = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"

    char = char.upper()
    assert char in alphabet, f"char must be in {alphabet}, but was {char}"

    height = alphabet.index(char) + 1
    width = alphabet.index(char) * 2 + 1

    for line in range(height):
        n_char = line * 2 + 1
        use_char = alphabet[line]
        line_string = use_char * n_char
        print(line_string.center(width))
```

```python
pyramid("h")
```
