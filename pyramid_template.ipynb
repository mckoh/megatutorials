# Aufgabe: Pyramid

**Schwierigkeitsgrad:** Mittel, 30 Punkte

**Ziel:** Entwickle eine Funktion pyramid(), die auf der Konsole eine symmetrische Pyramide bestehend aus Buchstaben und Leerzeichen ausgibt. Die Funktion soll einen beliebigen Buchstaben des Alphabets ("A" bis "Z") als Argument entgegennehmen und diesen anschließend zur Konstruktion der Pyramidenform verwenden. Beachte dabei folgende Regeln.

>**Hinweis:** Diese Aufgabe kann dir helfen die Aufgabe 'Diamond (Mittel, 30 Punkte)' zu lösen.

## Anforderungen
* Die erste Zeile enthält zentriert den Buchstaben A.
* Jede nachfolgende Zeile fügt symmetrisch zur vorherigen Zeile denselben Buchstaben hinzu.
* Alle Zeilen haben immer dieselbe Anzahl an Leerzeichen am Anfang und am Ende der Zeile, um die Pyramidenform zu bewahren.
* Die Pyramide ist horizontal symmetrisch.
* Die Buchstaben sind aufsteigend sortiert (A-Z) von oben nach unten.
* Jede Zeile enthält mehr Buchstaben als die vorherige, beginnend mit einem Buchstaben in der ersten Zeile, drei Buchstaben in der zweiten, und so weiter.
* Die Pyramidenhöhe entspricht der Position des Buchstabens im Alphabet (z.B. 'C' ergibt eine Pyramide mit 3 Zeilen).


## Testfälle
```python
pyramid('A')
# A

pyramid('C')
#     A    
#    BBB   
#   CCCCC  

pyramid('E')
#       A      
#      BBB     
#     CCCCC    
#    DDDDDDD   
#   EEEEEEEEE  
```
